```python
import pandas as pd
%matplotlib notebook
import matplotlib as mpl
import matplotlib.pyplot as plt
confirmed=pd.read_csv('time_series_covid19_confirmed_global.csv')
del confirmed['Province/State']
del confirmed['Lat']
del confirmed['Long']
confirmed_egypt=confirmed[confirmed['Country/Region']=='Egypt']
confirmed_Italy=confirmed[confirmed['Country/Region']=='Italy']
del confirmed_egypt['Country/Region']
del confirmed_Italy['Country/Region']
confirmed_egypt=confirmed_egypt.T
confirmed_Italy=confirmed_Italy.T
confirmed_Italy.columns=['total cases']
confirmed_egypt.columns=['total cases']
confirmed_Italy
```

```python
recovered=pd.read_csv('time_series_covid19_recovered_global.csv')
del recovered['Province/State']
del recovered['Lat']
del recovered['Long']
recovere_Egypt=recovered[recovered['Country/Region']=='Egypt']
recovere_Italy=recovered[recovered['Country/Region']=='Italy']
del recovere_Egypt['Country/Region']
del recovere_Italy['Country/Region']
recovere_Egypt=recovere_Egypt.T
recovere_Italy=recovere_Italy.T
recovere_Egypt.columns=['total recovere']
recovere_Italy.columns=['total recovere']
recovere_Italy
```

```python
deaths=pd.read_csv('time_series_covid19_deaths_global.csv')
deaths_Egypt=deaths[deaths['Country/Region']=='Egypt']
deaths_Italy=deaths[deaths['Country/Region']=='Italy']
del deaths_Egypt['Country/Region']
del deaths_Egypt['Province/State']
del deaths_Egypt['Lat']
del deaths_Egypt['Long']


del deaths_Italy['Country/Region']
del deaths_Italy['Province/State']
del deaths_Italy['Lat']
del deaths_Italy['Long']

deaths_Egypt=deaths_Egypt.T
deaths_Italy=deaths_Italy.T
deaths_Egypt.columns=['total deaths']
deaths_Italy.columns=['total deaths']
deaths_Italy
```

```python
import numpy as np
total_out=pd.merge(deaths_Egypt, recovere_Egypt, how='outer', left_index=True, right_index=True)
total_out=pd.merge(total_out, confirmed_egypt, how='outer', left_index=True, right_index=True)
total_out['toltal out']=total_out['total recovere']+total_out['total deaths']
date= np.arange('2020-01-22','2020-06-21', dtype='datetime64[D]')
total_out['current_cases']=total_out['total cases']-total_out['toltal out']


total_out_Italy=pd.merge(deaths_Italy, recovere_Italy, how='outer', left_index=True, right_index=True)
total_out_Italy=pd.merge(total_out_Italy, confirmed_Italy, how='outer', left_index=True, right_index=True)
total_out_Italy['toltal out']=total_out_Italy['total recovere']+total_out_Italy['total deaths']


date= np.arange('2020-01-22','2020-06-21', dtype='datetime64[D]')
total_out_Italy['current_cases']=total_out_Italy['total cases']-total_out_Italy['toltal out']

total_out_Italy
```

```python
plt.figure(figsize=(12,4))
ax1=plt.subplot(1, 2, 1)
y=total_out['current_cases'].values
plt.plot(date,y,'')
plt.plot(date,total_out['total deaths'].values,'r')
plt.plot(date,total_out['total recovere'].values,'g')
plt.title('Egypt covid 19 report')
plt.xlabel('Date')
#plt.Ylabel('people number')
plt.legend(['current cases','total deaths','total recovere'] )
ax2=plt.subplot(1, 2, 2 , sharey=ax1)
y2=total_out_Italy['current_cases'].values
plt.plot(date,y2,'')
plt.plot(date,total_out_Italy['total deaths'].values,'r')
plt.plot(date,total_out_Italy['total recovere'].values,'g')
plt.title('Italy covid 19 report')
plt.xlabel('Date')
plt.legend(['current cases','total deaths','total recovere'] )
#plt.Ylabel('people number')
plt.legend('current cases','total deaths' )
```
